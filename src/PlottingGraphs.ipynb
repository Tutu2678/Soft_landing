```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read data from CSV files
file1 = 'model/PER_DDQ_DDDQ.csv'  # Replace with your actual file names
file2 = 'model/mean_rewards_ppo.csv'
```

```python
# Load the data into Pandas DataFrames
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)
```

```python
x_label = 'Episode'  # Adjust this label as necessary
y_label = 'Reward'   # Adjust this label as necessary

# Plot graph
plt.figure(figsize=(10, 6))

plt.plot(df2['PPO'], label='PPO', color='blue')
plt.plot(df1['DQN'], label='DQN', color='green')
plt.plot( df1['DDQN'], label='DDQN', color='red')
plt.plot(df1['Priority'], label='PER', color='orange')

plt.xlabel(x_label)
plt.ylabel(y_label)
plt.title('Comparison of Models Performance')
plt.legend()
plt.grid(True)
plt.show()
```
